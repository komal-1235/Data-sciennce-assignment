# Task 1: Data Cleaning ang Preparation

```python
import pandas as pd
import numpy as np
```

## Import Dataset

```python
df = pd.read_csv("Cardiotocographic.csv")
```

## Missing values

```python
print(df.isnull().sum())
```

## Handle missing values using imputation

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
df_filled = pd.DataFrame(imputer.fit_transform(df),columns=df.columns)
```

### Check data type

```python
print(df_filled.dtypes)
```

## Task 2: Statistical Summary

### Calculate summary statistics

```python
summary = df_filled.describe()
print(summary)
```

```python
#Highlight interesting findings
print("Mean Baseline Fetal Heart Rate:", summary['LB']['mean'])
print("Standard Deviation of Acceleration:", summary['AC']['std'])
```

## Task 3: Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Histograms

```python
sns.histplot(df_filled['LB'])
plt.title('Baseline Fetal Heart Rate Distribution')
plt.show()
```

## Boxplots

```python
sns.boxplot(df_filled['AC'])
plt.title('Accelerations Distribution')
plt.show()
```

## Scatter plots

```python
sns.scatterplot(x='LB', y='AC', data=df_filled)
plt.title('Relationship between Baseline Fetal Heart Rate and Accelerations')
plt.show()
```

## Correlation heatmap

```python
corr = df_filled.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

## Task 4: Pattern Recognition and Insights

```python
# Identify correlations between variables
print(corr['LB'].sort_values(ascending=False))
```
