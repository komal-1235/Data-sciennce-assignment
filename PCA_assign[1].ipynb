## PCA 

```python
pca = PCA()
principalComponents = pca.fit_transform(X)
```

```python
principalComponents
```

```python
PCA_dataset = pd.DataFrame(data = principalComponents, columns = ['component1', 'component2', 'component3', 'component4', 
                                                                  'component5', 'component6','component7', 'component8', 'component9',
                                                                 'component10', 'component11', 'component12', 'component13'] )
PCA_dataset.head()
```

```python
var = pca.explained_variance_ratio_
var
```

```python
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
pca.components_
```

```python
plt.plot(var1,color="red")
```

## 3 principal components 

```python
principal_component1 = PCA_dataset['component1']
principal_component2 = PCA_dataset['component2']
principal_component3 = PCA_dataset['component3']
```

```python
pca_df = pd.concat([principal_component1, principal_component2, principal_component3], axis = 1)
pca_df.head()
```

```python
fig=plt.figure(figsize=(14,8))
sns.scatterplot(data=pca_df)
```

## KMeans

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
scaled = scalar.fit_transform(pca_df)
pd.DataFrame(scaled).describe()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2,init='random',n_init='auto')
kmeans.fit(scaled)
```

```python
kmeans.inertia_
```

```python
SSE = []
for cluster in range(1, 11):
    kmeans = KMeans(n_clusters=cluster,init='random',n_init='auto')
    kmeans.fit(pca_df)
    SSE.append(kmeans.inertia_)
    
plt.plot(range(1, 11), SSE)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()
```

```python
Y = KMeans(n_clusters = 4, init='random',n_init = 'auto')
Y.fit(scaled)
```

```python
Y.labels_
```

```python
wine['clustersid']=Y.labels_
wine1
```

```python
wine1['clustersid'].value_counts()
```

## Hierarchical 

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(10,8))
dendrogram=sch.dendrogram(sch.linkage(X,'complete'))
```

```python
h=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
h
```

```python
y=pd.DataFrame(h.fit_predict(X),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
wine['clustersid']=h.labels_
wine
```
